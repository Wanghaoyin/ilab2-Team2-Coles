```python
import json
import os
import pandas as pd
```

```python
def get_day_list(month):
    large_month_list = [1,3,5,7,8,10,12]
    
    if month in large_month_list:
        day_number = 31
    else:
        day_number = 30
    
    day_list = list(range(1,day_number+1))
    for i in range(0,day_number):
        if day_list[i] < 10:
            day_list[i] = '0' + str(day_list[i])
        else:
            day_list[i] = str(day_list[i])
            
    return day_list
```
